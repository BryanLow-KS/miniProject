## create custom images

```python
import matplotlib.pyplot as plt
```

```python
## create a black greyscale image
black = np.zeros((200, 100), dtype = np.uint8) # "uint8"

plt.imshow(black, cmap = plt.cm.gray) #'gray'
plt.title('black image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#create a white image 

white = np.zeros((200, 100)) + 255
white = np.uint8(white)
# white = white.astype('uint8')

plt.imshow(white, cmap = plt.cm.gray, vmin = 0, vmax = 256) #'gray'
plt.title('white image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# ramdom color image
random_color_img = np.random.randint(0, high = 256, size=(150, 250, 3), dtype=np.uint8)

plt.imshow(random_color_img) 
plt.title('random color image')
plt.xticks([]), plt.yticks([])
plt.show()
```

Q1. Try to write a simple custom function to determine whether a given integer is odd or even number.

```python
def OddEven(x): 
    if isinstance(x,int): 
        if x % 2 == 0: 
            if x == 0: 
                statement = ' is neither odd nor even.'
            else:
                statement = ' is even number.'
        else: 
            statement = ' is odd number.'
    else: 
        statement = " is not integer!"
        
    return print(f"{x}" + statement)
```

```python
OddEven(0)
```

```python
OddEven(3.3)
```

```python
OddEven(201)
```

```python
OddEven(10)
```

Q2. Write a simple example code to show that Numpy is more efficient in numerical computation of large arrays of data than equivalent Python list.

```python
large_n = 1e6 #1 million
numpy_arr = np.arange(large_n)
list_arr = list(range(int(large_n)))
```

```python
print(numpy_arr[:5])
print(list_arr[:5])
```

```python
print(numpy_arr[-5:])
print(list_arr[-5:])
```

```python
%timeit res = 2*numpy_arr
```

```python
%timeit res = [2*i for i in list_arr]
```

```python
#3. Run the following codes

my_arr = np.arange(10)
print(my_arr)
my_arr[4:7] = 25
print(my_arr)
arr_slice = my_arr[4:7]

# Change the first element of arr_slice to -1
arr_slice[0]= -1
 
print(arr_slice)
print(my_arr)

#What do you notice? Propose a way to reassign arr_slice with new value without modifying my_arr.
#The origical array my_arr is also modified, arr_slice is pointing directly to my_arr value, 
#therefore changes to arr_slice affects my_arr
```

```python
my_arr = np.arange(10)
print(my_arr)
my_arr[4:7] = 25
print(my_arr)
arr_slice = my_arr[4:7].copy() #make another copy to prevent directly pointing to the original array

# Change the first element of arr_slice to -1
arr_slice[0]= -1
 
print(arr_slice)
print(my_arr)
```

4. Create an image as shown as the following with the help of Numpy and matplotlib modules. You can arbitrarily set the dimension of the image and white circular spot at the middle. 



```python
import numpy as np
import matplotlib.pyplot as plt

figure, axes = plt.subplots()

image = np.zeros((200, 200))

Drawing_uncolored_circle = plt.Circle( (100,100), 20, fill = True, color = 'white')
axes.set_aspect( 1 )
axes.add_artist( Drawing_uncolored_circle )
    
plt.imshow(image, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
